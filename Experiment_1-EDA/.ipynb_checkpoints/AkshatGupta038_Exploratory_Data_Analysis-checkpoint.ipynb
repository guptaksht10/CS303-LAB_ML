Recap - what do each of these things do???


- df.shape:
Returns the dimensions of the DataFrame
- df.head():
Displays the first 5 rows by default.
- df.dtypes:
Shows the data type of each column
- df.describe():
 - Numeric:
Gives summary statistics for numerical columns by default.
 - Non-Numeric:
shows count, unique values, top (most frequent), frequency.
- df['column'].value_counts():
Shows unique values & their frequencies in a column.
- df.isnull().sum():
Counts the number of missing (NaN) values per column.
- df.isnull().sum().sum():
Counts the total number of missing values in the entire DataFrame.
- df.drop():
Drops rows or columns.

# [Objective 03](#pandas-visualizations) Describe and discriminate between basic data types

## Overview

One of the cornerstones of Exploratory Data Analysis (EDA) is being able to identify variable types.  We will need different statistical methods to display and describe each of these different types of data.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

The objective of this task is to perform eda on the Iris dataset to study the relationship between numerical features (Sepal Length, Sepal Width, Petal Length, Petal Width) and the categorical variable Species.

```python
iris_data = pd.read_csv('./dataset/iris_dirty.csv')
```

```python
iris_data.info()
```

```python
iris_data.describe()
```

```python
iris_data['Species'].value_counts()
```

```python
# same category written in different cases (setosa, Setosa, SETOSA etc.), which pandas treats as separate values.
iris_data['Species'] = iris_data['Species'].str.lower()
iris_data['Species'].value_counts()
```

```python
# Dropping unnamed col to perform EDA
iris_data = iris_data.drop(columns=['Unnamed: 0'])
```

```python
sns.set(style="whitegrid")
```

```python
#Univariate Plots
num_cols = iris_data.select_dtypes(include=['float64', 'int64']).columns  

plt.figure(figsize=(12, 8))

for i, col in enumerate(num_cols, 1): 
    plt.subplot(2, 2, i)  
    sns.histplot(iris_data[col], kde=True, bins=20, color="skyblue")
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")

plt.tight_layout()
plt.show()
```

All numerical features (sepal length, sepal width, petal length, petal width) are continuous.

Petal length & petal width show clear separation between species (multimodal distribution).

Sepal width looks more normally distributed but with a slight skew.

Sepal length has some spread and overlap across species.

```python
plt.figure(figsize=(6,4))
sns.barplot(x = 'Species', y="Sepal.Width", data=iris_data, estimator="mean", palette="coolwarm", hue='Species')
plt.title("Average Sepal Length per Species")
plt.xlabel("Species")
plt.ylabel("Average Sepal Length (cm)")
plt.show()
```

```python
plt.figure(figsize=(12, 8))

for i, col in enumerate(num_cols, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(y=iris_data[col], color="lightgreen")
    plt.title(f"Boxplot of {col}")
    plt.ylabel(col)

plt.tight_layout()
plt.show()
```

```python
sns.pairplot(iris_data, vars=num_cols, hue="Species", diag_kind="kde", palette="Set2")
plt.suptitle("Pairwise Scatterplots by Species", y=1.02)
plt.show()
```

Petal features: Strong separation among species, minimal overlap.

Sepal features: Overlap present, but Setosa still distinguishable.

Suggests that petal measurements are the most discriminative for ML models.

```python
plt.figure(figsize=(12,8))

for i,col in enumerate(num_cols, 1):
    plt.subplot(2, 2, i)  
    sns.boxplot(x="Species", y=col, data=iris_data, palette="pastel", hue='Species')
    plt.title(f"{col} Distribution by Species")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(iris_data[num_cols].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```
